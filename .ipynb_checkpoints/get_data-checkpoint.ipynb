```python
# Import
import pandas as pd
import numpy as np
import hvplot.pandas
from pathlib import Path
import os
from dotenv import load_dotenv
load_dotenv()
import csv
from binance import Client
import path
# Setting these options will allow for reviewing more of the DataFrames
pd.set_option('display.max_rows', 2000)
pd.set_option('display.max_columns', 2000)
pd.set_option('display.width', 1000)
```

```python
API_KEY = os.getenv('API_KEY')
API_SECRET = os.getenv('API_SECRET')

client = Client(API_KEY, API_SECRET)

csvfile= open('xrp1h.csv','w', newline='')

candlestick_writer = csv.writer(csvfile, delimiter=',')

candlesticks = client.get_historical_klines("XRPUSDT", Client.KLINE_INTERVAL_1HOUR, "20 aug 2017", "5 jan 2022")

for candlestick in candlesticks:
    candlestick[0] = candlestick[0] / 1000
    candlestick_writer.writerow(candlestick)
```

```python
csvfile.close()
```

```python
# CSV file coverter

btc1h= pd.read_csv(
Path("./data/btc1h.csv"))
btc1h = btc1h.iloc[:,[1,2,3,4,5]]
btc1h.columns =(['open','high','low','close','volume'])

date = pd.date_range(start='20/AUG/2017', end='5/JAN/2022', freq='1H')
date = pd.DataFrame(date)
btc1h = btc1h.assign(Date=date)
btc1h.set_index(['Date'],inplace = True)
btc1h.tail(10)
```

```python
btc1h.to_csv('BTC1H.csv')
```
